# Word-Clouds

Word Clouds are a way of displaying the importance of words in a collection of texts. The more frequent the word is, the greater space it occupies in the image. This notebook page examines the corpus, taking into account only some specific parts of speech.

For extracting the sentiment from a collection of words, not every part of speech has the same weight. For example adjectives, more specifically prepositive adjectives are often used by people to express an emotion by describing the followed noun in a statement. Nouns on the other hand are the part of speech used to name things like creatures, objects, places, actions, qualities, states of existence, or ideas. In natural language processing the word used to describe or call an object or product can be a key feature, especially for a classification problem.

The image at the top reveals that the most found adjectives in the reviews of this dataset are "good", "much" and "great" which are irrefutably positive adjectives. A good point to mention is that the entire meaning of those words changes by adding the word "not" ahead of them, so in a classification problem the pair of tokens which includes the word "not" could be considered as one another token, for example "notgood", in order to differ from the regular word "good". In addition, the word "bad", "old" and "little" are those implying negative sentiment. A few other words like "original", "different" and "hard" can be considered words with neutral sentiment or words that need to be accompanied to other words in order to express a feeling.
